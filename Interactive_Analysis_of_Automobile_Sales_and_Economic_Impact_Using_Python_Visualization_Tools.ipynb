```python
import pandas as pd
import matplotlib.pyplot as plt

# ✅ Sample dummy dataset for demonstration
data = {
    'Year': [2006, 2007, 2008, 2009, 2010, 2011]*3,
    'Vehicle_Type': ['Sedan']*6 + ['SUV']*6 + ['Truck']*6,
    'Sales': [50000, 52000, 48000, 45000, 47000, 49000,
              60000, 63000, 58000, 54000, 56000, 58000,
              40000, 42000, 38000, 35000, 36000, 37000]
}

# Create DataFrame
df = pd.DataFrame(data)

# ✅ Add Recession indicator (assuming 2008–2009 were recession years)
df['Recession'] = df['Year'].apply(lambda x: 1 if x in [2008, 2009] else 0)

# ✅ Filter only recession period data
recession_df = df[df['Recession'] == 1]

# ✅ Pivot table: index = Year, columns = Vehicle_Type, values = Sales
pivot_table = recession_df.pivot_table(values='Sales', index='Year', columns='Vehicle_Type', aggfunc='sum')

# ✅ Plotting
plt.figure(figsize=(10, 6))
for column in pivot_table.columns:
    plt.plot(pivot_table.index, pivot_table[column], marker='o', label=column)

plt.title('Vehicle Sales Trends by Vehicle Type During Recession Periods')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.legend(title='Vehicle Type')
plt.tight_layout()

# ✅ Save plot
plt.savefig('Line_plot_2.png')
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ✅ Sample dummy dataset
data = {
    'Year': [2006, 2007, 2008, 2009, 2010, 2011]*3,
    'Vehicle_Type': ['Sedan']*6 + ['SUV']*6 + ['Truck']*6,
    'Sales': [50000, 52000, 48000, 45000, 47000, 49000,
              60000, 63000, 58000, 54000, 56000, 58000,
              40000, 42000, 38000, 35000, 36000, 37000]
}

# ✅ Create DataFrame
df = pd.DataFrame(data)

# ✅ Label recession years (2008 and 2009)
df['Period'] = df['Year'].apply(lambda x: 'Recession' if x in [2008, 2009] else 'Non-Recession')

# ✅ Plotting with Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Year', y='Sales', hue='Vehicle_Type', style='Period', markers=True, dashes=False)

plt.title('Sales Trends per Vehicle Type (Recession vs. Non-Recession)')
plt.ylabel('Sales')
plt.grid(True)
plt.legend(title='Vehicle Type / Period', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# ✅ Save plot
plt.savefig('Bar_Chart.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ✅ Sample dataset (replace with actual data if available)
data = {
    'Year': list(range(2006, 2012)),
    'GDP': [14.5, 14.8, 14.2, 13.9, 14.3, 14.7],
    'Recession': [False, False, True, True, False, False]
}

# ✅ Create DataFrame
df = pd.DataFrame(data)

# ✅ Separate data into recession and non-recession
recession_df = df[df['Recession'] == True]
non_recession_df = df[df['Recession'] == False]

# ✅ Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5), sharey=True)

# Plot for Recession Period
axes[0].plot(recession_df['Year'], recession_df['GDP'], marker='o', color='red')
axes[0].set_title('GDP During Recession')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('GDP (Trillions USD)')
axes[0].grid(True)

# Plot for Non-Recession Period
axes[1].plot(non_recession_df['Year'], non_recession_df['GDP'], marker='o', color='green')
axes[1].set_title('GDP During Non-Recession')
axes[1].set_xlabel('Year')
axes[1].grid(True)

plt.tight_layout()
plt.savefig('Subplot.png')  # 📌 Required filename for Coursera submission
plt.show()
```

```python
from dash import Dash, html, dcc, Input, Output

# Initialize app
app = Dash(__name__)
app.title = "Automobile Sales Dashboard"

# App layout
app.layout = html.Div([
    html.H1("📊 Automobile Sales Dashboard",
            style={'textAlign': 'center', 'color': '#003366', 'marginTop': '20px'}),

    html.P("An interactive dashboard to analyze automobile sales and economic factors.",
           style={'textAlign': 'center', 'fontSize': '18px', 'color': '#555555'}),

    html.Div([
        html.Label("Select Report Type:", style={'fontWeight': 'bold'}),
        dcc.Dropdown(
            id='report-type-dropdown',
            options=[
                {'label': 'Recession Report', 'value': 'recession'},
                {'label': 'Yearly Report', 'value': 'yearly'}
            ],
            value='recession',
            placeholder="Select a report type"
        )
    ], style={'width': '50%', 'margin': 'auto', 'marginTop': '30px'}),

    html.Div([
        html.Label("Select Vehicle Type:", style={'fontWeight': 'bold'}),
        dcc.Dropdown(
            id='vehicle-type-dropdown',
            options=[
                {'label': 'Sedan', 'value': 'sedan'},
                {'label': 'SUV', 'value': 'suv'},
                {'label': 'Truck', 'value': 'truck'},
                {'label': 'Sports Car', 'value': 'sports'}
            ],
            value='sedan',
            placeholder="Select a vehicle type"
        )
    ], style={'width': '50%', 'margin': 'auto', 'marginTop': '20px'}),

    # Output container
    html.Div(
        id='output-container',
        className='output-box',
        children=[],
        style={
            'marginTop': '40px',
            'textAlign': 'center',
            'padding': '20px',
            'border': '2px solid #cccccc',
            'borderRadius': '10px',
            'width': '80%',
            'margin': 'auto'
        }
    )
])

# ✅ Callback function to update output based on dropdown selections
@app.callback(
    Output('output-container', 'children'),
    Input('report-type-dropdown', 'value'),
    Input('vehicle-type-dropdown', 'value')
)
def update_output(report_type, vehicle_type):
    if not report_type or not vehicle_type:
        return html.Div("Please select both a report type and a vehicle type.")
    return html.Div([
        html.H3(f"You selected: {report_type.title()} Report for {vehicle_type.title()} Vehicles"),
        html.P("This section will display relevant analysis and visualizations based on the selected inputs.")
    ])

# Run the Dash app
if __name__ == '__main__':
    app.run(debug=True)
```

```python
import pandas as pd
import plotly.express as px
from dash import Dash, html, dcc

# Load the dataset
df = pd.read_csv("extracted_data/Auto Sales data.csv", parse_dates=['ORDERDATE'], dayfirst=True)

# Group data by date for total sales
sales_by_date = df.groupby("ORDERDATE").sum(numeric_only=True).reset_index()

# Line chart: Total Sales Over Time
fig_sales_trend = px.line(
    sales_by_date,
    x="ORDERDATE", y="SALES",
    title="Total Automobile Sales Over Time",
    labels={"SALES": "Total Sales", "ORDERDATE": "Order Date"}
)

# Bar chart: Sales by Product Line
sales_by_product = df.groupby("PRODUCTLINE").sum(numeric_only=True).reset_index()
fig_product_sales = px.bar(
    sales_by_product,
    x="PRODUCTLINE", y="SALES",
    title="Sales by Product Line",
    labels={"SALES": "Total Sales", "PRODUCTLINE": "Product Line"}
)

# Pie chart: Deal Size Distribution
deal_size_distribution = df["DEALSIZE"].value_counts().reset_index()
deal_size_distribution.columns = ["DEALSIZE", "Count"]
fig_deal_size_pie = px.pie(
    deal_size_distribution,
    names="DEALSIZE", values="Count",
    title="Deal Size Distribution"
)

# Dash App
app = Dash(__name__)
app.title = "Auto Sales Dashboard"

# Layout
app.layout = html.Div([
    html.H1("📊 Automobile Sales Dashboard", style={'textAlign': 'center'}),

    dcc.Graph(figure=fig_sales_trend),
    dcc.Graph(figure=fig_product_sales),
    dcc.Graph(figure=fig_deal_size_pie)
])

# Run the server
if __name__ == '__main__':
    app.run(debug=True)
```

```python
import pandas as pd
import plotly.express as px
from dash import Dash, html, dcc

# Load the dataset
df = pd.read_csv("extracted_data/Auto Sales data.csv", parse_dates=['ORDERDATE'], dayfirst=True)

# Extract Year from ORDERDATE
df["Year"] = df["ORDERDATE"].dt.year

# Yearly Total Sales
yearly_sales = df.groupby("Year").sum(numeric_only=True).reset_index()
fig_yearly_sales = px.line(
    yearly_sales, x="Year", y="SALES",
    title="📈 Total Automobile Sales Per Year",
    labels={"SALES": "Total Sales"}
)

# Yearly Sales by Product Line
yearly_product_sales = df.groupby(["Year", "PRODUCTLINE"]).sum(numeric_only=True).reset_index()
fig_yearly_product_sales = px.bar(
    yearly_product_sales, x="Year", y="SALES", color="PRODUCTLINE",
    title="📊 Sales by Product Line Per Year",
    barmode='group',
    labels={"SALES": "Sales", "PRODUCTLINE": "Product Line"}
)

# Dash App
app = Dash(__name__)
app.title = "Yearly Report Statistics"

app.layout = html.Div([
    html.H1("📆 Yearly Report Statistics", style={'textAlign': 'center'}),

    html.Div([
        dcc.Graph(figure=fig_yearly_sales),
        dcc.Graph(figure=fig_yearly_product_sales),
    ], style={'width': '90%', 'margin': 'auto'})
])

if __name__ == '__main__':
    app.run(debug=True)
```
